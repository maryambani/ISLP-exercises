1. Describe the null hypotheses to which the p-values given in Table 3.4
correspond. Explain what conclusions you can draw based on these
p-values. Your explanation should be phrased in terms of sales, TV,
radio, and newspaper, rather than in terms of the coefficients of the
linear model.

In table 3.4, the null hypotheses would be the following: 
* Intercept: Expected sales would be zero when no money is spent on radio, TV and newspaper advertising.
* TV: TV advertising budget has no effect on the number of units sold
* Radio: Radio advertising budget has no effect on the number of units sold.
* Newspaper: No relationship between newspaper advertising and the number of units sold.

Based on the p values, we can reject the null hypotheses for the intercept, TV and radio because they have p-values close to zero. We cannot reject the null hypothesis for newspaper since the p value is bigger than the significance level (0.05). 
This means that newspaper advertising indeed has no effect on the number of units sold, but that TV and radio advertising will increase the amount of sales (positive coefficient indicating a positive relationship between both variables). Rejecting the intercept null hypothesis also means that we can expect some sales even when no advertising money is spent through either means of TV, radio or newspaper.

2. Carefully explain the differences between the KNN classifier and KNN
regression methods.

KNN will estimate the probability of class j as a fraction and will assign the class with the highest estimated probability to the new test observation
With KNN Classifiers you have decision boundaries that depend on the value of K that you choose. 

K = 1 : Overly flexible boundry, overly sensitive to noise (high variance, low bias)

K=100 : Less flexible, decision boundary looks more linear (Low variance, high bias)

But KNN can also be used for regression
KNN regression is a non parametric method closely related to KNN classifier 
first identifies K training observations (neighbors) closest to xo. It estimates f(x0) using average of all training responses. 

1) Pick a value for K / number of neighbors to consider
2) Calculate the distances between the training points and new point (Euclidean distance)
3) Find K training points that are the closest 
4) Calculate the average of these values of the K neighbors
5) Assign the average value as the prediction for the new point

KNN regression assigns an average value as the prediction to the new data point instead of KNN classification.

Difference between both:

1) KNN Classification will give a fraction to predict a category (class label) to a new observation.
    KNN Regression on the other hand has continuous predictions that are quantitative for the new observation 
2) KNN Classifier uses the majority vote to assign the most common class label among neighbors to the new observation, or calculates the class probabilities depending on the proportion of each class of the neighbors.
    KNN Regression uses an average of the target values of the neighbors and assigns the mean to the new data point : all neighbors equally contribute to the final prediction.
    
    Equations:
    
    KNN Classifier:
    $$ 
    Pr(Y = j \mid X = x_0) = \frac{1}{K} \sum\limits_{i  \in  \mathcal{N}_0} I(yᵢ = j)
    $$
    Where I(yᵢ = j) is an indicator function that equals 1 if the neighbor belongs to class j, otherwise 0. 
    
    KNN Regression:

    $$
    f̂(x_0) = \frac{1}{K} \sum\limits_{i \in \mathcal{N}_0} yᵢ
    $$
    Where yᵢ is the continuous target value of neighbor i
    
    $\mathcal{N}_0 $: K training observations closest to $x_0$
